```python
import pandas as pd
from county import counties, states, duplicates, uniques
import matplotlib.pyplot as plt
```

```python
file = 'combined_states_agesex.csv'

df= pd.read_csv(file)
```

```python
df
```

```python
add = []

for n in duplicates:
    temp_df = df.loc[df["CTYNAME"]==n[0], :]
    temp_df = temp_df.loc[df["STNAME"]==n[1], :]
    add.append(temp_df)
```

```python
for n in uniques:
    temp_df = df.loc[df["CTYNAME"]==n, :]
    add.append(temp_df)
lime_df = pd.concat(add)
lime_df
```

```python
breakdown = list(lime_df.columns)
breakdown.pop(0)
print(len(breakdown))
```

```python
overall = []
for s in breakdown:
    if s[len(s)-3:len(s)] == 'TOT':
        overall.append(s)
overall
```

```python
lis = ['AGE16PLUS_TOT','AGE18PLUS_TOT']

new = []
for n in overall:
    if n not in lis:
        new.append(n)

less_detail = new[0:8]
more_detail = new[8:len(new)-1]
print(less_detail)
print(more_detail)
#new
```
