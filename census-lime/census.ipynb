```python
import pandas as pd
from county import counties, states, duplicates, uniques
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = 'acs2017_county_data.csv'

df = pd.read_csv(file)

df.head()
```

```python
lime_df = df.loc[df["County"].isin(counties), :]
lime_df
```

```python
add = []

for n in duplicates:
    temp_df = lime_df.loc[df["County"]==n[0], :]
    temp_df = temp_df.loc[df["State"]==n[1], :]
    add.append(temp_df)
```

```python
for n in uniques:
    temp_df = lime_df.loc[df["County"]==n, :]
    add.append(temp_df)
clean_df = pd.concat(add)
clean_df
```

```python
#For all of USA
men_percentage = df['Men'].sum()/df['TotalPop'].sum()
print(f"Total percentage of men is {men_percentage*100}%")
print(f"Total percentage of women is {(1-men_percentage)*100}%")

white_population = (df['TotalPop']*df['White']/100).sum()
white_per = (white_population/df['TotalPop'].sum())
print(f"The percentage of white people is {white_per*100}%")

asian_population = (df['TotalPop']*df['Asian']/100).sum()
asian_per = (asian_population/df['TotalPop'].sum())
print(f"The percentage of Asian people is {asian_per*100}%")

asian_population = (df['TotalPop']*df['Black']/100).sum()
asian_per = (asian_population/df['TotalPop'].sum())
print(f"The percentage of Black people is {asian_per*100}%")

asian_population = (df['TotalPop']*df['Hispanic']/100).sum()
asian_per = (asian_population/df['TotalPop'].sum())
print(f"The percentage of Hispanic people is {asian_per*100}%")

voting_age = df['VotingAgeCitizen'].sum()/df['TotalPop'].sum()
print(f"The percentage of Voting Age People is {voting_age*100}%")

asian_population = (df['TotalPop']*df['Walk']/100).sum()
asian_per = (asian_population/df['TotalPop'].sum())
print(f"The percentage of people who walk to work is {asian_per*100}%")


income = (df['TotalPop']*df['IncomePerCap']).sum()/df['TotalPop'].sum()
print(f"The average income per capita is ${income}")
```

```python
#For Specific Counties
men_percentage = clean_df['Men'].sum()/clean_df['TotalPop'].sum()
print(f"Total percentage of men is {men_percentage*100}%")
print(f"Total percentage of women is {(1-men_percentage)*100}%")

white_population = (clean_df['TotalPop']*clean_df['White']/100).sum()
white_per = (white_population/clean_df['TotalPop'].sum())
print(f"The percentage of white people is {white_per*100}%")

asian_population = (clean_df['TotalPop']*clean_df['Asian']/100).sum()
asian_per = (asian_population/clean_df['TotalPop'].sum())
print(f"The percentage of Asian people is {asian_per*100}%")

asian_population = (clean_df['TotalPop']*clean_df['Black']/100).sum()
asian_per = (asian_population/clean_df['TotalPop'].sum())
print(f"The percentage of Black people is {asian_per*100}%")

asian_population = (clean_df['TotalPop']*clean_df['Hispanic']/100).sum()
asian_per = (asian_population/clean_df['TotalPop'].sum())
print(f"The percentage of Hispanic people is {asian_per*100}%")

voting_age = clean_df['VotingAgeCitizen'].sum()/clean_df['TotalPop'].sum()
print(f"The percentage of Voting Age People is {voting_age*100}%")

asian_population = (clean_df['TotalPop']*clean_df['Walk']/100).sum()
asian_per = (asian_population/clean_df['TotalPop'].sum())
print(f"The percentage of people who walk to work is {asian_per*100}%")

income = (clean_df['TotalPop']*clean_df['IncomePerCap']).sum()/clean_df['TotalPop'].sum()
print(f"The average income per capita is ${income}")
```

```python
#For all counties
vals = df['TotalPop'].tolist()
plt.boxplot(vals)
plt.xlabel('County Populations')
plt.ylabel('Population (tens millions)')
plt.show()
```

```python
nums = [.25,.5,.75]
all_quarts = []
for n in nums:
    all_quarts.append(np.quantile(vals, n))
all_quarts
```

```python
for x in all_quarts:
    print(x/10000000)
```

```python
iqr = all_quarts[2]- all_quarts[0]

upper = all_quarts[2]+(1.5*iqr)
lower = all_quarts[0]-(1.5*iqr)
print(f'The US lower bound is {lower}')
print(f'The US upper bound is {upper}')
```

```python
#For lime counties
values = clean_df['TotalPop'].tolist()
plt.boxplot(values)
plt.xlabel('County Populations')
plt.ylabel('Population (millions)')
plt.show()
```

```python
iqr_lime = lime_quarts[2] -lime_quarts[0]

lime_upper = lime_quarts[2]+(1.5*iqr_lime)
lime_lower = lime_quarts[0]-(1.5*iqr_lime)
print(f'The lime lower bound is {lime_lower}')
print(f'The lime upper bound is {lime_upper}')
```
